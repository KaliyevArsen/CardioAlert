```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report
from xgboost import XGBClassifier
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
column_names = [
    'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
    'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target'
]

data = pd.read_csv('dataset/processed.cleveland.data', header=None, names=column_names)

# ? -> NaN
data = data.replace('?', np.nan)

# Преобразуем столбцы к числовому типу
for col in data.columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Исключим строки с полностью пустыми значениями (если есть)
data = data.dropna(axis=0, how='all')

# Целевая: если target > 0, значит есть болезнь, иначе нет
data['target'] = data['target'].apply(lambda x: 1 if x > 0 else 0)

X = data.drop('target', axis=1)
y = data['target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
X = data.drop('target', axis=1)
y = data['target']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('clf', XGBClassifier(eval_metric='logloss', random_state=42))
])

param_grid = {
    'clf__n_estimators': [100, 200, 300],
    'clf__max_depth': [2, 3, 4],
    'clf__learning_rate': [0.01, 0.05, 0.1],
    'clf__reg_alpha': [0, 0.1, 0.5],
    'clf__reg_lambda': [1, 5, 10]
}

grid_search = GridSearchCV(pipeline, param_grid, scoring='roc_auc', cv=10, n_jobs=-1)
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
print("Лучшие параметры:", grid_search.best_params_)
```

```python
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:, 1]

print("Точность (Accuracy):", accuracy_score(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_proba))
print("Отчёт классификации:\n", classification_report(y_test, y_pred))
```

```python
joblib.dump(best_model, 'model.pkl')
print("Модель сохранена в model.pkl")
```

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(
    best_model, X, y, cv=5, scoring='roc_auc', n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 5)
)

train_means = train_scores.mean(axis=1)
train_stds = train_scores.std(axis=1)
test_means = test_scores.mean(axis=1)
test_stds = test_scores.std(axis=1)

plt.figure(figsize=(8,5))
plt.plot(train_sizes, train_means, 'o-', color='r', label='Обучающая выборка')
plt.plot(train_sizes, test_means, 'o-', color='g', label='Валидация')
plt.fill_between(train_sizes, train_means - train_stds, train_means + train_stds, color='r', alpha=0.1)
plt.fill_between(train_sizes, test_means - test_stds, test_means + test_stds, color='g', alpha=0.1)
plt.title('Кривые обучения')
plt.xlabel('Размер обучающей выборки')
plt.ylabel('ROC-AUC')
plt.legend(loc='best')
plt.show()
```

```python
from sklearn.model_selection import validation_curve

param_range = [1, 2, 3, 4, 5]
train_scores, test_scores = validation_curve(
    best_model['clf'], X, y, param_name="max_depth", param_range=param_range,
    cv=5, scoring="roc_auc", n_jobs=-1
)

train_means = train_scores.mean(axis=1)
test_means = test_scores.mean(axis=1)
train_stds = train_scores.std(axis=1)
test_stds = test_scores.std(axis=1)

plt.figure(figsize=(8,5))
plt.plot(param_range, train_means, 'o-', color='r', label='Обучение')
plt.plot(param_range, test_means, 'o-', color='g', label='Валидация')
plt.fill_between(param_range, train_means - train_stds, train_means + train_stds, alpha=0.1, color='r')
plt.fill_between(param_range, test_means - test_stds, test_means + test_stds, alpha=0.1, color='g')
plt.title("Валидационная кривая по max_depth")
plt.xlabel("Max Depth")
plt.ylabel("ROC-AUC")
plt.legend(loc="best")
plt.show()
```

```python
importances = best_model.named_steps['clf'].feature_importances_
feature_names = X.columns
feat_imp = pd.Series(importances, index=feature_names).sort_values(ascending=False)
plt.figure(figsize=(8,6))
sns.barplot(x=feat_imp, y=feat_imp.index, palette='viridis')
plt.title("Важность признаков")
plt.xlabel("Значимость")
plt.ylabel("Признак")
plt.show()
```
